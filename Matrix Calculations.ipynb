Trying to Calculate the Covariance Matrix Here, not so sure

```python
array = np.array([data["y1"],data["y2"],data["y3"],data["y4"]]).astype(float)    #astype(float) averts the error 'float' object has no attribute 'shape'
cov_matrix = np.cov(array)                                                       #the calculations are unbiased (N-1) but they can be set to biased (N)

cov_matrix
```

Getting the inverse of the 4X4 Covariance Matrix M

```python
# M = np.array([[6.013475,5.879625,5.48815,5.27075],[5.879625,6.126875,5.84575,5.62725],[5.48815,5.84575,6.5721,6.599],[5.27075,5.62725,6.599,7.0915]]) from excel
M = np.array(cov_matrix)
M_inverse = np.linalg.inv(M)
M_inverse
```

Calculating the determinant

```python
M_det = np.linalg.det(M).round(4)
M_det
```
